```python
import pandas as pd

df = pd.read_csv("GB180925-JO_Extended_NPX.csv")
# --- Step 2: Filter out Olink controls (rows 2-289) ---
# Assuming first row is row 0
df_target = df.iloc[289:].copy()  # row 290 onward
# keep only SampleID starting with "RB01_"
df_target = df_target[df_target['SampleID'].str.startswith("RB01_")]
# Pivot Count and PCNormalizedNPX separately ---

# Count OP
df_count = df_target.pivot(index='SampleID', columns='Assay', values='Count')

# PCNormalizedNPX OP
df_npx = df_target.pivot(index='SampleID', columns='Assay', values='PCNormalizedNPX')

# --- Save to CSV ---
df_count.to_csv("proteins_count_wide.csv")
df_npx.to_csv("proteins_PCNormalizedNPX_wide.csv")

print("Files saved: proteins_count_wide.csv and proteins_PCNormalizedNPX_wide.csv")
```

```python
import pandas as pd

# Read data
df = pd.read_csv("proteins_PCNormalizedNPX_wide.csv")

# Extract ID and time label from SampleID
df[['ID', 'TimeLabel']] = df['SampleID'].str.extract(r'^(RB\d+_\d+)\s+(.*)$')

# Create numeric code for timepoints
mapping = {'Baseline': 0, 'D30': 2, 'Post': 1}
df['Timepoint'] = df['TimeLabel'].map(mapping)

# Clean SampleID: remove everything after the first space
df['SampleID'] = df['SampleID'].str.split().str[0]

# Reorder columns: place Timepoint right after SampleID
cols = df.columns.tolist()
cols.insert(cols.index('SampleID') + 1, cols.pop(cols.index('Timepoint')))
df = df[cols]

# Drop helper columns if not needed
df = df.drop(columns=['ID', 'TimeLabel'])

# Save new file
df.to_csv("NPXpp_with_timecode.csv", index=False)

print("Files saved")
```
